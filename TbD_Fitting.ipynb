```python
##This is for HF TC parameter caluclation
## Sung Chung

import scipy as sp
import numpy as np
from numpy import linspace,exp
import matplotlib.pyplot as plt
import pandas as pd
import math
from pandas.tools.plotting import andrews_curves
from scipy import interpolate
%matplotlib inline
pd.options.display.max_columns = 100


peakAtNativeFld=-1        # if HF peak is at negative field +1, or -1(Tbd HF is -1)  
num_dev=0
absMaxFld=13900            #****may need to slightly cut end of the testing field
fltLeft=-3000*peakAtNativeFld
fltRight=4000*peakAtNativeFld
##################
directory= "C:\python"
filename='y5ghmc12sd 51-100_result_Norm'
####################
filepath=directory+"\\"+filename+".csv"
#____________________________________
DataFile=filename + "_DataTable.csv"
sumName=filename + "_Summary.csv"
sumf=directory+"\\"+sumName
dataf=directory+"\\"+DataFile

dfb=pd.read_csv(filepath,index_col = [0]).transpose()   #Add index_col = [0] to not have index values as own row during transposition
dfb=dfb.dropna()

volt=[1.9,2,2.1,2.2,2.3,2.4,2.5,2.55,2.6,2.65,2.7,2.75,2.8,2.85,2.9,2.95,3,3.05,3.1,3.15,3.2,3.25,3.3,3.35,3.4]
dfb['V']=volt


dfb.tail()
```

```python
fitDeg=6
outName=filename + "_Fitdegree_"+ str(fitDeg)+".csv"
outputF=directory+"\\"+outName

colName=dfb.columns.values
devN=[]
crtV=[]
PartID=[]

testedN=dfb.columns.size-1

for i in range(0,testedN):
    PartID.append(colName[i])    
    
    x1=dfb.loc[:,"V"]
    y1=dfb.iloc[:,i]   
    
    fp=sp.polyfit(x1,y1,4)
    f=sp.poly1d(fp)
    plt.plot(x1,y1,'o',x1,f(x1))
    
    
    fphalf=sp.polyfit(x1,y1-0.5,fitDeg)
    fhalf=sp.poly1d(fphalf)
    
    
    
    roots=np.roots(fhalf)
#     dev[i-1]=i
    
    for k in roots:
        if ((k.real>1.9 and k.real<3.4) and k.imag==0):
            devN.append(i)
            crtV.append(k.real)


dfT=pd.DataFrame({"devN":devN, "Critical Voltage(V)":crtV,'Fitdeg':fitDeg, "PartID":PartID })
dfT.head()
dfT.to_csv(outputF)
    

plt.title(outName)
plt.axhline(y=0.5, color='r', linestyle='-')
plt.show()
```
